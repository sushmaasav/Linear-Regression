## Multiple Linear Regression Model

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_multi,y,test_size = 0.2, random_state=1)
print("X_train shape: ",X_train.shape)
print("X_test shape: ",X_test.shape)
print("y_train shape: ",y_train.shape)
print("y_test shape: ",y_test.shape)
```

```python
MLR3 = LinearRegression()
```

```python
MLR3.fit(X_train,y_train)
```

```python
y_predict_test = MLR3.predict(X_test)
```

```python
y_predict_train = MLR3.predict(X_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_predict_test,y_test)
```

```python
r2_score(y_predict_train,y_train)
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge
```

```python
lm_ridge = Ridge(alpha = 0.5)
```

```python
lm_ridge.fit(X_train,y_train)
```

```python
r2_score(y_test,lm_ridge.predict(X_test))
```

```python
from sklearn.model_selection import validation_curve
```

```python
param_range = np.logspace(-2,8,100)
```

```python
train_scores, test_scores = validation_curve(Ridge(), X_train,y_train, param_name="alpha",param_range= param_range, scoring = 'r2')
```

```python
print(train_scores)
print(test_scores)
```

```python
train_mean = (np.mean(train_scores,axis=1))
test_mean = (np.mean(test_scores,axis=1))
```

```python
max(train_mean)
```

```python
max(test_mean)
```

```python
sns.jointplot(x = np.log(param_range),y=test_mean)
```

```python
np.where(test_mean == max(test_mean))
```

```python
param_range[20]
```

```python
lm_r_best = Ridge(alpha = param_range[20])
```

```python
lm_r_best.fit(X_train,y_train)
```

```python
r2_score(y_test, lm_r_best.predict(X_test))
```

```python
r2_score(y_train,lm_r_best.predict(X_train))
```

```python
lm_r_best.coef_
```

```python
lm_r_best.intercept_
```
